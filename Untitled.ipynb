```python
# =========================
# Image Augmentation Agent
# =========================
# –ó–∞–ø—É—Å–∫–∞—Ç—å –≤ Google Colab
# pip install openai pillow

import base64
import json
from pathlib import Path
from io import BytesIO
from PIL import Image

from openai import OpenAI
# from google.colab import files


# =========================
# 0. –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∏–µ–Ω—Ç–∞
# =========================

client = OpenAI()


# =========================
# 1. –ó–∞–≥—Ä—É–∑–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
# =========================

print("–ó–∞–≥—Ä—É–∑–∏ —Ñ–æ—Ç–æ —Ç–æ–≤–∞—Ä–∞:")

# uploaded = files.upload()
# image_path = next(iter(uploaded))

image_path = "gen1200 (3).jpg"
print(f"–ó–∞–≥—Ä—É–∂–µ–Ω–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ: {image_path}")


# =========================
# 2. –ó–∞–ø—Ä–æ—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
# =========================

user_prompt = "–¥–æ–±–∞–≤—å –Ω–∞ –∫–∞–ø–æ—Ç –¥–µ–≤—É—à–∫—É —Å –ø–ª—è–∂–∞, –∫–∞–∫ –±—É–¥—Ç–æ —Å–∏–¥–∏—Ç –Ω–∞ –Ω–µ–º, —Å—Ü–µ–Ω–∞ —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–∞—è"

if not user_prompt:
    raise ValueError("–ó–∞–ø—Ä–æ—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø—É—Å—Ç.")


# =========================
# 3. LLM: Domain / Realism / Mode
# =========================

mode_response = client.responses.create(
    model="gpt-4.1-mini",
    input=f"""
–¢—ã –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—à—å –∑–∞–ø—Ä–æ—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫ –∞–≥–µ–Ω—Ç—É –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–π.

–¢–≤–æ—è –∑–∞–¥–∞—á–∞ ‚Äî –≤—ã–ø–æ–ª–Ω–∏—Ç—å –¢–†–ò –ø—Ä–æ–≤–µ—Ä–∫–∏.

‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
1. DOMAIN CHECK
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
–ó–∞–ø—Ä–æ—Å –¥–æ–ø—É—Å—Ç–∏–º –¢–û–õ–¨–ö–û –µ—Å–ª–∏ –æ–Ω –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∫ –∞–≤—Ç–æ–º–æ–±–∏–ª—é
–∏ –≤–∏–∑—É–∞–ª—å–Ω–æ–º—É –∏–∑–º–µ–Ω–µ–Ω–∏—é —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–∏ –º–∞—à–∏–Ω—ã.

‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
2. REALISM CHECK
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
–î–æ–ø—É—Å—Ç–∏–º—ã –¢–û–õ–¨–ö–û —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã–µ, —Ñ–∏–∑–∏—á–µ—Å–∫–∏ –≤–æ–∑–º–æ–∂–Ω—ã–µ —Å—Ü–µ–Ω—ã.

–†–ê–ó–†–ï–®–ï–ù–û:
- –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤ (—á–µ–º–æ–¥–∞–Ω, –∫–æ—Ñ–µ, —Å—É–º–∫–∞, –∫–∞–ª—å—è–Ω)
- –ø—Ä–µ–¥–º–µ—Ç—ã –º–æ–≥—É—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –ù–ê –∏–ª–∏ –†–Ø–î–û–ú —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º
- —Å—Ü–µ–Ω–∞ –≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∫ –æ–±—ã—á–Ω–∞—è —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏—è

–ó–ê–ü–†–ï–©–ï–ù–û:
- —Ñ–∞–Ω—Ç–∞–∑–∏–π–Ω—ã–µ —Å—Ü–µ–Ω—ã
- —Ö—É–¥–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Ä–µ–Ω–¥–µ—Ä
- –∏–ª–ª—é—Å—Ç—Ä–∞—Ü–∏—è
- –∏–≥—Ä—É—à–µ—á–Ω—ã–π –º–∞—Å—à—Ç–∞–±
- –∏–∑–º–µ–Ω–µ–Ω–∏–µ –≥–µ–æ–º–µ—Ç—Ä–∏–∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è
- —Å–º–µ–Ω–∞ –ø–æ–≥–æ–¥—ã, –≤—Ä–µ–º–µ–Ω–∏ —Å—É—Ç–æ–∫, –æ–∫—Ä—É–∂–µ–Ω–∏—è

‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
3. MODE DETECTION
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
–û–ø—Ä–µ–¥–µ–ª–∏ —Ä–µ–∂–∏–º:

- "improve"  ‚Äî —É–ª—É—á—à–µ–Ω–∏–µ –∫–∞—á–µ—Å—Ç–≤–∞ —Ñ–æ—Ç–æ –ë–ï–ó –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –æ–±—ä–µ–∫—Ç–æ–≤
- "augment"  ‚Äî –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –û–î–ù–û–ì–û –æ–±—ä–µ–∫—Ç–∞ –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Å—Ü–µ–Ω—ã

–û—Ç–≤–µ—Ç—å –°–¢–†–û–ì–û –≤ —Ñ–æ—Ä–º–∞—Ç–µ JSON –±–µ–∑ –ø–æ—è—Å–Ω–µ–Ω–∏–π:

{{
  "domain": "car" | "not_car",
  "realism": "acceptable" | "unacceptable",
  "mode": "improve" | "augment"
}}

–ó–∞–ø—Ä–æ—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è:
{user_prompt}

–í–æ–∑–≤—Ä–∞—â–∞–π –¢–û–õ–¨–ö–û –æ–¥–∏–Ω JSON-–æ–±—ä–µ–∫—Ç.
–ù–µ –¥–æ–±–∞–≤–ª—è–π –Ω–∏–∫–∞–∫–æ–π —Ç–µ–∫—Å—Ç –¥–æ –∏–ª–∏ –ø–æ—Å–ª–µ JSON.
"""
)

def extract_json_from_response(response):
    """
    –ù–∞–¥—ë–∂–Ω–æ–µ –∏–∑–≤–ª–µ—á–µ–Ω–∏–µ JSON –∏–∑ Responses API.
    –ï—Å–ª–∏ JSON –Ω–µ –Ω–∞–π–¥–µ–Ω ‚Äî –∫–∏–¥–∞–µ—Ç –ø–æ–Ω—è—Ç–Ω—É—é –æ—à–∏–±–∫—É.
    """

    if not hasattr(response, "output"):
        raise ValueError("–û—Ç–≤–µ—Ç LLM –Ω–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –ø–æ–ª—è output.")

    collected_text = []

    for message in response.output:
        if getattr(message, "type", None) == "message":
            for content in getattr(message, "content", []):
                if getattr(content, "type", None) == "output_text":
                    text = getattr(content, "text", "").strip()
                    if text:
                        collected_text.append(text)

    if not collected_text:
        # üëá –ö–†–ò–¢–ò–ß–ï–°–ö–ò –í–ê–ñ–ù–û –¥–ª—è –¥–µ–±–∞–≥–∞
        raise ValueError(
            "LLM –Ω–µ –≤–µ—Ä–Ω—É–ª–∞ output_text.\n"
            f"RAW RESPONSE:\n{response}"
        )

    raw_text = "\n".join(collected_text)

    try:
        return json.loads(raw_text)
    except json.JSONDecodeError as e:
        raise ValueError(
            "LLM –≤–µ—Ä–Ω—É–ª–∞ –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π JSON.\n"
            f"–¢–µ–∫—Å—Ç:\n{raw_text}"
        ) from e



analysis = extract_json_from_response(mode_response)

# =========================
# 4. –í–∞–ª–∏–¥–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
# =========================

if analysis["domain"] != "car":
    raise ValueError("–ó–∞–ø—Ä–æ—Å –æ—Ç–∫–ª–æ–Ω—ë–Ω: –Ω–µ –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∫ –∞–≤—Ç–æ–º–æ–±–∏–ª—é.")

if analysis["realism"] != "acceptable":
    raise ValueError("–ó–∞–ø—Ä–æ—Å –æ—Ç–∫–ª–æ–Ω—ë–Ω: –Ω–µ—Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–∞—è —Å—Ü–µ–Ω–∞.")

mode = analysis["mode"]

if mode not in {"improve", "augment"}:
    raise ValueError("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Ä–µ–∂–∏–º –æ–±—Ä–∞–±–æ—Ç–∫–∏.")

print(f"–†–µ–∂–∏–º –æ–ø—Ä–µ–¥–µ–ª—ë–Ω: {mode}")


# =========================
# 5. –§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ image prompt
# =========================

if mode == "improve":
    image_prompt = f"""
–¢—ã ‚Äî –∞–≥–µ–Ω—Ç —É–ª—É—á—à–µ–Ω–∏—è —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–π –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–π –¥–ª—è –æ–±—ä—è–≤–ª–µ–Ω–∏–π.

–ó–ê–î–ê–ß–ê:
–£–ª—É—á—à–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –Ω–∞ –æ—Å–Ω–æ–≤–µ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ —Ñ–æ—Ç–æ.

–°–¢–†–û–ì–ò–ï –ü–†–ê–í–ò–õ–ê:
- –ò—Å–ø–æ–ª—å–∑—É–π –¢–û–õ–¨–ö–û –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
- –ù–ï –¥–æ–±–∞–≤–ª—è–π –Ω–æ–≤—ã–µ –æ–±—ä–µ–∫—Ç—ã
- –ù–ï –º–µ–Ω—è–π —Ñ–æ—Ä–º—É, —Ü–≤–µ—Ç –∏ –≥–µ–æ–º–µ—Ç—Ä–∏—é –∞–≤—Ç–æ–º–æ–±–∏–ª—è
- –ù–ï —Å–∫—Ä—ã–≤–∞–π –∏ –Ω–µ –º–∞—Å–∫–∏—Ä—É–π –¥–µ—Ñ–µ–∫—Ç—ã
- –†–∞–∑—Ä–µ—à–µ–Ω–æ:
  ‚Ä¢ —É–ª—É—á—à–∏—Ç—å —Ä–µ–∑–∫–æ—Å—Ç—å
  ‚Ä¢ —É–ª—É—á—à–∏—Ç—å —ç–∫—Å–ø–æ–∑–∏—Ü–∏—é
  ‚Ä¢ —Å–ª–µ–≥–∫–∞ —É–ª—É—á—à–∏—Ç—å –∫–æ–Ω—Ç—Ä–∞—Å—Ç
- –ó–∞–ø—Ä–µ—â–µ–Ω—ã —Ö—É–¥–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —Å—Ç–∏–ª–∏ –∏ —Ä–µ–Ω–¥–µ—Ä
- –ò—Ç–æ–≥ –¥–æ–ª–∂–µ–Ω –≤—ã–≥–ª—è–¥–µ—Ç—å –∫–∞–∫ —Ä–µ–∞–ª—å–Ω–æ–µ —Ñ–æ—Ç–æ

–ó–ê–ü–†–û–° –ü–û–õ–¨–ó–û–í–ê–¢–ï–õ–Ø:
{user_prompt}
"""

else:  # augment
    image_prompt = f"""
–¢—ã ‚Äî –∞–≥–µ–Ω—Ç –ª–æ–∫–∞–ª—å–Ω–æ–≥–æ –¥–æ–ø–æ–ª–Ω–µ–Ω–∏—è —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è.

–ó–ê–î–ê–ß–ê:
–î–æ–±–∞–≤–∏—Ç—å –û–î–ò–ù —Ñ–∏–∑–∏—á–µ—Å–∫–∏ –≤–æ–∑–º–æ–∂–Ω—ã–π –æ–±—ä–µ–∫—Ç
–Ω–∞ –∏—Å—Ö–æ–¥–Ω–æ–µ —Ñ–æ—Ç–æ –∞–≤—Ç–æ–º–æ–±–∏–ª—è.

–°–¢–†–û–ì–ò–ï –ü–†–ê–í–ò–õ–ê (–û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–´):
- –ò—Å—Ö–æ–¥–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –¥–æ–ª–∂–Ω–æ –æ—Å—Ç–∞—Ç—å—Å—è –§–û–¢–û–ì–†–ê–§–ò–ï–ô
- –ó–∞–ø—Ä–µ—â–µ–Ω–æ –º–µ–Ω—è—Ç—å:
  ‚Ä¢ –æ—Å–≤–µ—â–µ–Ω–∏–µ
  ‚Ä¢ —Ü–≤–µ—Ç–æ–≤—É—é —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—É
  ‚Ä¢ –ø–æ–≥–æ–¥—É
  ‚Ä¢ —Ñ–æ–Ω
  ‚Ä¢ —Å—Ç–∏–ª—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
- –ó–∞–ø—Ä–µ—â–µ–Ω–æ –ø–µ—Ä–µ—Ä–∏—Å–æ–≤—ã–≤–∞—Ç—å –∞–≤—Ç–æ–º–æ–±–∏–ª—å
- –ó–∞–ø—Ä–µ—â–µ–Ω–æ —É–ª—É—á—à–∞—Ç—å —Å—Ü–µ–Ω—É
- –ó–∞–ø—Ä–µ—â–µ–Ω—ã –∫–∏–Ω–æ—ç—Ñ—Ñ–µ–∫—Ç—ã –∏ —Ö—É–¥–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Å—Ç–∏–ª—å

–†–ê–ó–†–ï–®–ï–ù–û:
- –î–æ–±–∞–≤–∏—Ç—å –¢–û–õ–¨–ö–û –æ–¥–∏–Ω –æ–±—ä–µ–∫—Ç
- –û–±—ä–µ–∫—Ç –¥–æ–ª–∂–µ–Ω –≤—ã–≥–ª—è–¥–µ—Ç—å –∫–∞–∫ —Ä–µ–∞–ª—å–Ω–æ –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–π
- –ú–∞—Å—à—Ç–∞–±, –ø–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–∞ –∏ —Ç–µ–Ω–∏ ‚Äî —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã–µ
- –í—Å—ë –¥–æ–ª–∂–Ω–æ –≤—ã–≥–ª—è–¥–µ—Ç—å –∫–∞–∫ –æ–±—ã—á–Ω–æ–µ —Ñ–æ—Ç–æ —Å —Ç–µ–ª–µ—Ñ–æ–Ω–∞

–û–ë–™–ï–ö–¢ –î–õ–Ø –î–û–ë–ê–í–õ–ï–ù–ò–Ø:
{user_prompt}

–í–ê–ñ–ù–û:
–ï—Å–ª–∏ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å –æ–±—ä–µ–∫—Ç –±–µ–∑ –ø–µ—Ä–µ—Ä–∏—Å–æ–≤–∫–∏ —Å—Ü–µ–Ω—ã ‚Äî
–≤—ã–ø–æ–ª–Ω–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤–º–µ—à–∞—Ç–µ–ª—å—Å—Ç–≤–æ –∏ —Å–æ—Ö—Ä–∞–Ω–∏ —Ñ–æ—Ç–æ.
"""


# =========================
# 6. –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
# =========================

with Image.open(image_path) as img:
    img = img.convert("RGB")
    buffer = BytesIO()
    img.save(buffer, format="JPEG", quality=95)
    buffer.seek(0)
    buffer.name = "input.jpg"


# =========================
# 7. Image-to-image edit
# =========================

result = client.images.edit(
    model="gpt-image-1",
    image=buffer,
    prompt=image_prompt,
    size="1024x1024"
)


# =========================
# 8. –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
# =========================

image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)

Path("result.png").write_bytes(image_bytes)

print("–ì–æ—Ç–æ–≤–æ. –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ —Ñ–∞–π–ª result.png")
```
